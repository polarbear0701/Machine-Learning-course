```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
bostonHouseFrame = pd.read_csv('housing-2.data.csv', delimiter='\\s+')
```

```python
bostonHouseFrame
```

```python
pd.DataFrame.min(bostonHouseFrame)
```

```python
pd.DataFrame.max(bostonHouseFrame)
```

```python
pd.DataFrame.mean(bostonHouseFrame)
```

```python
pd.DataFrame.median(bostonHouseFrame)
```

```python
plt.figure()
bostonHouseFrame.hist()
plt.show()
```

```python
bostonHouseFrame.plot(kind='box', subplots=True, layout=(4,4), sharex=False, sharey=True)
plt.show()
```

```python
correlations = bostonHouseFrame.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1, cmap=plt.cm.PuBu)
fig.colorbar(cax)
ticks = np.arange(0,14,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(bostonHouseFrame.columns)
ax.set_yticklabels(bostonHouseFrame.columns)
plt.show()
```
